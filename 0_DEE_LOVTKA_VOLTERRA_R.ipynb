```r
# http://tbb.bio.uu.nl/rdb/grindR/
# Este script ilustra ejemplos de simulacin, manipulacin y ajuste de modelos dinmicos usando grindR.
# Se incluyen ejemplos de ODEs, manejo de eventos, mapas discretos, ecuaciones con retardo, modelos con mltiples ecuaciones
# y ajuste de parmetros.

# ------------------------------
# Configuracin inicial
# ------------------------------

# (Opcional) Limpiar el entorno de trabajo
# rm(list=ls())

# Mostrar el directorio de trabajo actual
getwd()

# Cargar el archivo de funciones Grind.R (asegrate de que est en el mismo directorio o especifica la ruta)
source('Grind.R')

# ======================================================
# 1. Modelo Predador-Presa (Ecuaciones Diferenciales Ordinarias)
# ======================================================

# Definicin del modelo predador-presa
# Variables:
#   P: poblacin de presas
#   D: poblacin de depredadores
# Parmetros:
#   alpha, beta, gamma, delta
model <- function(t, state, parms) {
  with(as.list(c(state, parms)), {
    dP <- alpha * P - beta * P * D       # Crecimiento de presas y depredacin
    dD <- gamma * P * D - delta * D        # Crecimiento de depredadores a partir de la depredacin
    return(list(c(dP, dD)))
  })
}

# Definir parmetros y condiciones iniciales
p <- c(alpha = 2/3, beta = 4/3, gamma = 1, delta = 1)
s <- c(P = 1, D = 1)

# Simulacin bsica del modelo con grfica de P y D
run(xlab = "Time", ylab = "Predator / Prey", ymax = 2, tmax = 25, tstep = 0.01)

# ------------------------------
# Manejo de eventos en la simulacin
# ------------------------------

# Ejemplo 1: Al llegar a t = 20 se fija la poblacin de depredadores (D) a 0
# Se utiliza "D" en lugar de "N" para que coincida con las condiciones iniciales
run(after = "if(t==20) state[\"D\"] <- 0", 
    legend = FALSE, tstep = 0.01, 
    xlab = "Time", ylab = "Predator / Prey", ymax = 5, tmax = 25)

# Ejemplo 2: Modificar aleatoriamente un parmetro (en este caso 'r')
# Nota: 'r' no est definido en este modelo; es solo un ejemplo ilustrativo.
run(after = "parms[\"r\"] <- rnorm(1, 1, 0.1)", legend = FALSE)

# Ejemplo 3: Otra forma de modificar el parmetro usando la posicin en el vector
run(after = "parms[1] <- rnorm(1, 1, 0.1)")

# Ejemplo 4: De nuevo, al llegar a t = 20 se fija la poblacin de depredadores (D) a 0
run(after = "if(t==20) state[\"D\"] <- 0")

# Uso de 'arrest' para detener la simulacin en un tiempo especfico y aplicar un evento
# Se corrige la variable de "N" a "D" para mantener consistencia
run(50, arrest = 33.14, after = "if(t==33.14) state[\"D\"] <- 0", table = TRUE)

# ------------------------------
# Aadir ruido gaussiano a las variables
# ------------------------------

# Se parte de un estado de equilibrio obtenido con newton()
# Se deben usar los mismos nombres de variables: P y D
f <- newton(c(P = 0.5, D = 0.5))
run(state = f, after = "state <- state + rnorm(2, mean = 0, sd = 0.01)")
# rnorm(2, mean=0, sd=0.01) genera dos valores aleatorios para P y D, respectivamente

# ------------------------------
# Manejo de eventos usando funcin (similar a deSolve)
# ------------------------------

# Se define una funcin que, al ocurrir el evento, fija la poblacin de depredadores (D) a 0
fun <- function(t, y, parms) {
  y["D"] <- 0
  return(y)
}
run(events = list(func = fun, time = 20))

# ------------------------------
# Uso de 'tweak' para manipular la solucin numrica
# ------------------------------

# Se aade una nueva columna 'T' que es la suma de las dos variables (P y D)
run(tweak = "nsol$T <- nsol[,2] + nsol[,3]", timeplot = TRUE)
# Alternativamente, usando apply para generalizar la suma:
run(tweak = "nsol$T <- apply(nsol[,2:3], 1, sum)")

# ======================================================
# 2. Mapas (Sistemas Dinmicos Discretos)
# ======================================================

# En este ejemplo se estudia un mapa discreto usando integracin por Euler.
# Se define una nica ecuacin para la variable N.
model <- function(t, state, parms) {
  with(as.list(c(state, parms)), {
    dN <- r * N * (1 - N) - N   # Mapa logstico modificado
    return(list(c(dN)))
  })
}
p <- c(r = 3.75)
s <- c(N = 0.01)
# Se simula el sistema discreto por 1000 pasos
data <- run(1000, method = "euler", table = TRUE)
# Diagrama de dispersin: N[t] vs N[t+1] para visualizar la dinmica
plot(data$N[1:999], data$N[2:1000], pch = ".")

# ======================================================
# 3. Ecuaciones Diferenciales con Retardo (DDEs)
# ======================================================

# Modelo Lotka-Volterra con retardo en el crecimiento del depredador.
# Variables:
#   R: poblacin de presas
#   N: poblacin de depredadores
model <- function(t, state, parms) {
  with(as.list(c(state, parms)), {
    tlag <- t - Delta  # Tiempo retrasado
    if (tlag < 0) {
      lags <- c(0, 0)  # Antes de que se active el retardo, se asume 0
    } else {
      lags <- lagvalue(tlag)  # Obtiene los valores retrasados de R y N
    }
    dR <- r * R * (1 - R/K) - a * R * N
    dN <- a * lags[1] * lags[2] - d * N
    return(list(c(dR, dN)))
  })
}

# Definir parmetros y condiciones iniciales para el modelo con retardo
p <- c(r = 1, K = 1, a = 1, c = 1, d = 0.5, Delta = 10)
s <- c(R = 1, N = 0.1)
# Ejecutar la simulacin utilizando el solver para DDEs
run(delay = TRUE)

# ======================================================
# 4. Modelos con Vectores de Ecuaciones (Mltiples Especies)
# ======================================================

# Este ejemplo modela la competencia entre n especies de presas, cada una con su propio depredador.
# Se usan variables R1, R2, ..., Rn para presas y N1, N2, ..., Nn para depredadores.
model <- function(t, state, parms) {
  with(as.list(c(state, parms)), {
    R <- state[1:n]             # Extraer las poblaciones de presas
    N <- state[(n + 1):(2 * n)]   # Extraer las poblaciones de depredadores
    S <- sum(R)                 # Suma total de presas para el trmino logstico
    dR <- b * R * (1 - S) - d1 * R - a * R * N
    dN <- a * R * N - d2 * N
    return(list(c(dR, dN)))
  })
}

# Nmero de especies
n <- 3
# Parmetro b para cada especie (valores aleatorios)
b <- rnorm(n, mean = 1, sd = 0.1)
# Otros parmetros del modelo
p <- c(d1 = 0.1, d2 = 0.2, a = 1)
# Condiciones iniciales para presas (R) y depredadores (N)
R <- rep(0.1 / n, n)
names(R) <- paste("R", seq(1, n), sep = "")
N <- rep(0.01 / n, n)
names(N) <- paste("N", seq(1, n), sep = "")
# Combinar en un nico vector de estado
s <- c(R, N)
# Ejecutar la simulacin
run()

# ======================================================
# 5. Ajuste de Parmetros (Fitting)
# ======================================================

# Definir un modelo para el ajuste de parmetros basado en dinmica presa-depredador.
# Variables:
#   R: poblacin de presas
#   N: poblacin de depredadores
model <- function(t, state, parms) {
  with(as.list(c(state, parms)), {
    dR <- r * R * (1 - R/K) - a * R * N
    dN <- c * a * R * N - delta * N
    return(list(c(dR, dN)))
  })
}

# Parmetros nominales para el modelo de ajuste
p <- c(r = 1, K = 1, a = 1, c = 1, delta = 0.5)
# Condiciones iniciales
s <- c(R = 1, N = 0.01)

# Generar un conjunto de datos FAKE simulando el modelo por 20 unidades de tiempo
data <- run(20, table = TRUE)

# Generar conjeturas aleatorias para condiciones iniciales y parmetros
s <- s * abs(rnorm(2, 1, 0.1)); s   # Nueva conjetura para condiciones iniciales
p <- p * abs(rnorm(5, 1, 0.1)); p   # Nueva conjetura para parmetros

# Ajustar el modelo a los datos con todas las variables libres
f <- fit()
summary(f)  # Mostrar resumen del ajuste, intervalos de confianza, etc.

# Extraer algunos parmetros ajustados (por ejemplo, posiciones 3 a 7)
p <- f$par[3:7]; p

# Generar otra conjetura aleatoria para los parmetros
p <- p * abs(rnorm(5, 1, 0.1)); p
# Definir los nombres de las variables libres (condiciones iniciales y parmetros)
w <- c(names(s), names(p))
# Realizar el ajuste nuevamente usando 'w' para indicar las variables libres
f <- fit(data, free = w)

# Alternativamente, tomar las condiciones iniciales directamente desde los datos
f <- fit(data, initial = TRUE, free = names(p))

# Dividir el conjunto de datos en dos: uno con la variable R y otro con N
dataR <- data; dataR$N <- NULL
dataN <- data; dataN$R <- NULL
# Ajustar el modelo usando ambos conjuntos de datos (resultado equivalente)
f <- fit(list(dataR, dataN))

# Redefinir parmetros y condiciones para generar un nuevo conjunto de datos
p <- c(r = 1, K = 1, a = 1, c = 1, delta = 0.5)
p["K"] <- 0.75            # Cambiar el valor del parmetro K
s <- c(R = 0.5, N = 0.05)   # Nuevas condiciones iniciales
data2 <- run(25, table = TRUE)  # Generar un nuevo conjunto de datos
s <- c(R = 0.75, N = 0.02)       # Conjetura promedio para condiciones iniciales

# Configurar la ventana grfica para dos paneles (lado a lado)
par(mfrow = c(1, 2))

# Ajustar el modelo a ambos conjuntos de datos permitiendo diferencias en R, N y K
f <- fit(list(data, data2), differ = c("R", "N", "K"), main = c("A", "B"))
f$par  # Mostrar los parmetros ajustados

# Proveer conjeturas iniciales individuales como lista
differ <- list(R = c(0.9, 0.55), N = c(0.02, 0.04), K = c(1.1, 0.7))
f <- fit(list(data, data2), differ = differ, main = c("A", "B"))

# Definir parmetros fijos como lista
fixed <- list(R = c(1, 0.5), N = c(0.01, 0.05))
differ <- "K"              # Se ajusta solo el parmetro K
free <- names(p)[-2]       # Se ajustan los otros cuatro parmetros compartidos
f <- fit(list(data, data2), free = free, differ = differ, fixed = fixed, main = c("A", "B"))

# Otra alternativa: tomar condiciones iniciales desde los datos
f <- fit(list(data, data2), free = free, differ = differ, initial = TRUE, main = c("A", "B"))

# Realizar bootstrap con 100 simulaciones para evaluar la robustez del ajuste
f <- fit(list(data, data2), free = free, differ = differ, fixed = fixed, main = c("A", "B"), bootstrap = 100)$par

# Restaurar la configuracin grfica a una sola ventana
par(mfrow = c(1, 1))
```
